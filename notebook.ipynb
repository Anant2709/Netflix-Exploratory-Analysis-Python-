```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# FINDING WHAT THE MOST COMMON DURATION OF FILMS WAS DURING THE 1990S

# Filtering data frame for just movies in the 1990s
#boolean array depicting what is movie and what is not
movie_bool = netflix_df["type"] == "Movie"
netflix_movies = netflix_df[movie_bool]
#boolean array showing year < 2000
movie_year = (netflix_movies["release_year"] < 2000) & (netflix_movies["release_year"] > 1989)
movie_list = netflix_movies[movie_year]
# Taking just the duration column in consideration 
dur_col = movie_list["duration"]
# bins = max(dur_col) as it will give us how frequent every duration is
plt.hist(dur_col, bins = max(dur_col))
plt.xlabel("Movie Durations")
plt.ylabel("Frequency")
plt.title("Frequency distribution of Movie durations in the 1990s")
# shows the frequency distribution of all the movies' durations in thr 90s
plt.show() 

#fair estimate from the graph of the duration of max frequency
duration = 93
print("An estimate from the graph of the most frequent duration of a movie in the 1990s is " + str(duration))
```

```python
# COUNTING SHORT ACTION MOVIES IN COMPARISON TO ALL SHORT MOVIES

# Filtering movie_list dataframe for only short action movies
# Correcting the filter condition to ensure proper grouping
short_action_movies = netflix_df[(netflix_df["genre"] == "Action") & (netflix_df["duration"] < 90) & (netflix_df["type"] == "Movie")]
short_action_movie_count = len(short_action_movies)

# Counting all short movies
short_movies = netflix_df[(netflix_df["duration"] < 90) & (netflix_df["type"] == "Movie")]
short_movie_count = len(short_movies)

# Counting all action movies
action_movies = netflix_df[(netflix_df["genre"] == "Action") & (netflix_df["type"] == "Movie")]
action_movie_count = len(action_movies)

# Calculating the percentage of short action movies in the pool of all short movies and all action movies
percentage_short_action_movies = (short_action_movie_count / short_movie_count) * 100
percentage = (short_action_movie_count / action_movie_count) * 100
# Displaying the results
print("The Number of short action movies are " + str(short_action_movie_count))
print("The Number of all short movies are " + str(short_movie_count))
print("The percenatage of short action movies are from all short movies is " + str(percentage_short_action_movies))
print("The percentage of short movies against all action movies is " + str(percentage)) 
```
